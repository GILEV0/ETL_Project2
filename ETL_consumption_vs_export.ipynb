```python
import pandas as pd
```

```python
coffee_importdf = pd.read_csv("Resources/Coffee_import.csv")
coffee_reexportdf = pd.read_csv("Resources/Coffee_re_export.csv")
coffee_importdf
```

```python
coffee_mergedf = pd.merge(coffee_importdf,coffee_reexportdf, on="Country", how="inner")
coffee_mergedf
```

```python
coffee_df = coffee_mergedf[["Country","Total_import","Total_re_export"]]
coffee_df.set_index("Country")
```

```python
coffee_df["Import_vs_Export"]= coffee_df["Total_import"]-coffee_df["Total_re_export"]
coffee_df.set_index("Country")
```
